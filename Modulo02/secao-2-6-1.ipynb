```python
# Desenvolver um programa para solicitar o valor de uma temperatura
# em graus Celsius e convertê-la para grau Fahrenheit.
# A fórmula de conversão é:
# F = C * 9/5 + 32
```

```python
celsius = float(input("Digite a temperatura em Celsius: "))
fahrenheit = celsius * 9/5 + 32
print("A temperatura em Fahrenheit é: ", fahrenheit, "ºF", sep="")
```

Não se preocupe com imperfeições no código - tudo bem se ele aceitar um tempo inválido - o mais importante é que o código produz resultados válidos para dados de entrada válidos.
